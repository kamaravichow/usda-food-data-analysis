```python

class NutrientSearch:
    df_nutrients_details: pd.DataFrame
    df_nutrients: pd.DataFrame

    def __init__(
        self, df_nutrients_details: pd.DataFrame, df_nutrients: pd.DataFrame
    ) -> None:
        self.df_nutrients_details = df_nutrients_details
        self.df_nutrients = df_nutrients

    def search_nutrient(
        self, name: str, valid_nutrient: bool = True, n_foods: bool = True
    ) -> pd.DataFrame:
        """
        - valid_nutrient flag indicates to check whether the id
        exists in the column of df_nutrients.
        - n_foods flag calculates
        number of measurements in df_nutrientfor each search result.
        It takes longer if True.
        """
        df_result = self._raw_search(name)
        if valid_nutrient:
            df_result = self._valid_ids(df_result)
        if not n_foods:
            return df_result
        return self._add_n_foods(df_result)

    def get_all_nutrients(
        self, valid_nutrient: bool = True, n_foods: bool = True
    ) -> pd.DataFrame:
        """
        - valid_nutrient flag indicates to check whether the id
        exists in the column of df_nutrients.
        - n_foods flag calculates number of foods in df_nutrient
        for each search result. It takes longer if True.
        """
        cols = ["name", "unit_name"]
        df_result = self.df_nutrients_details[cols]
        if valid_nutrient:
            df_result = self._valid_ids(df_result)
        if not n_foods:
            return df_result
        return self._add_n_foods(df_result)

    # ----------------------------------------------------------------

    def _raw_search(self, string: str) -> pd.DataFrame:
        cols = ["name", "unit_name"]
        return self.df_nutrients_details[cols].loc[
            self.df_nutrients_details.name.str.contains(string, case=False)
        ]

    def _valid_ids(self, df_nut_details_res: pd.DataFrame) -> pd.DataFrame:
        df_result = self._valid_cols(df_nut_details_res)
        df_result = self._valid_non_nan(df_result)
        df_result.index.name = "id"
        return df_result

    def _valid_cols(self, df_nut_details_res: pd.DataFrame) -> pd.DataFrame:
        return df_nut_details_res.loc[
            df_nut_details_res.index.intersection(self.df_nutrients.columns)
        ]

    def _valid_non_nan(self, df_nut_details_res: pd.DataFrame) -> pd.DataFrame:
        return df_nut_details_res.loc[
            df_nut_details_res.index.intersection(
                self.df_nutrients[df_nut_details_res.index]
                .dropna(axis=1, how="all")
                .columns
            )
        ]

    def _add_n_foods(self, df_nut_details_res: pd.DataFrame) -> pd.DataFrame:
        result_idxs = df_nut_details_res.index.intersection(
            self.df_nutrients.columns
        )
        df_result = df_nut_details_res.copy().loc[result_idxs]
        df_result.loc[:, "n_foods"] = self._get_n_meas_for_ids(result_idxs)
        return df_result

    def _get_n_meas_for_ids(self, ids: list[int]) -> list[int]:
        return [self.df_nutrients[id_].dropna().shape[0] for id_ in ids]
```

```python
from typing import Optional

import pandas as pd


class FoodNameSearch:
    df_foods: pd.DataFrame
    df_branded_foods: pd.DataFrame
    df_nutrients: pd.DataFrame

    def __init__(
        self,
        df_foods: pd.DataFrame,
        df_branded_foods: pd.DataFrame,
        df_nutrients: pd.DataFrame,
    ) -> None:
        self.df_foods = df_foods
        self.df_branded_foods = df_branded_foods
        self.df_nutrients = df_nutrients

    def search_single_fdc_id(
        self, fdc_id: int, valid_nutrients: bool = True
    ) -> Optional[str]:
        """
        - valid_nutrients flag filters fdc_id search to include
        only foods with valid nutrient measuments.
        """
        nutrients_flag = valid_nutrients and not self._valid_nutrients_fdc_id(
            fdc_id
        )
        if (fdc_id not in self.df_foods.index) or nutrients_flag:
            return None
        name = self.df_foods.description[fdc_id]
        if not self._is_branded_food(fdc_id):
            return name
        return f"{name} | {self.df_branded_foods.brand_owner[fdc_id]}"

    def search_fdc_ids(
        self,
        fdc_ids: tuple[int, ...] | list[int] | pd.Series,
        valid_nutrients: bool = True,
    ) -> pd.Series:
        """
        resulting series is indexed by fdc_id.
        - valid_nutrients flag filters fdc_id search to include
        only foods with valid nutrient measuments.
        """
        result_idxs = pd.Index(fdc_ids).intersection(self.df_foods.index)
        if valid_nutrients:
            result_idxs = self._valid_nutrients_fdc_ids(result_idxs)
        return self._get_name_series(result_idxs)

    def search_by_name(
        self, name: str, valid_nutrients: bool = True
    ) -> pd.Series:
        """
        resulting series is indexed by fdc_id.
        - valid_nutrients flag filters fdc_id search to include
        only foods with valid nutrient measuments.
        """
        search_mask = self.df_foods.description.str.contains(name, case=False)
        result_idxs = self.df_foods.loc[search_mask.fillna(False)].index
        if valid_nutrients:
            result_idxs = self._valid_nutrients_fdc_ids(result_idxs)
        return self._get_name_series(result_idxs)

    # ----------------------------------------------------------------

    def _valid_nutrients_fdc_id(self, fdc_id: int) -> bool:
        return not self.df_nutrients.loc[fdc_id].dropna().empty

    def _valid_nutrients_fdc_ids(self, fdc_ids: pd.Index) -> pd.Index:
        # doing the intersection first is not necessary, however increases
        # performance
        fdc_idx = fdc_ids.intersection(self.df_nutrients.index)
        return fdc_ids.intersection(
            self.df_nutrients.loc[fdc_idx].dropna(axis=0, how="all").index
        )

    def _is_branded_food(self, fdc_id: int) -> bool:
        data_type = self.df_foods.data_type[fdc_id]
        if (
            data_type != "branded_food"
            or fdc_id not in self.df_branded_foods.index
        ):
            return False
        else:
            return True

    def _get_name_series(self, result_fdc_idxs: pd.Index) -> pd.Series:
        food_result = self.df_foods.description[result_fdc_idxs]
        # branded foods name disambiguation
        branded_foods_idxs = result_fdc_idxs.intersection(
            self.df_branded_foods.brand_owner.dropna().index
        )
        if not branded_foods_idxs.empty:
            branded_foods_names = self._get_branded_foods_names(
                food_result, branded_foods_idxs
            )
            food_result[branded_foods_idxs] = branded_foods_names
        food_result.name = "food_name"
        return food_result

    def _get_branded_foods_names(
        self, food_result: pd.Series, branded_foods_idxs: pd.Index
    ) -> pd.Series:
        """add brand owner string to branded_food name"""
        return (
            food_result[branded_foods_idxs]
            + " | "
            + self.df_branded_foods.brand_owner[branded_foods_idxs]
        )
```

```python
from typing import Any
```

```python


def get_maximal_no_nan_df(
    df: pd.DataFrame, min_data_size: int | float = 0.1
) -> pd.DataFrame:
    """
    -df has NaN values
    - min_data_size: if int >= 1 then it is the minimum number of rows
    -   to keep while looking for no NaN rows
    -   if float < 1 then the minimum number or rows to keep is
    -   min_data_size * len(df)
    """
    df_append, df_result = _setup_init_dfs(df)
    add_more_cols = True
    remaining_cols: list[Any] = df.columns.tolist()
    while add_more_cols:
        no_nan_n_rows = _get_no_nan_rows(df, df_append, remaining_cols)
        col_2_append = remaining_cols[np.argmax(no_nan_n_rows)]
        df_append = _append_2_df_append(df_append, df, col_2_append)
        remaining_cols.remove(col_2_append)
        add_more_cols = _should_add_more_cols(df_append, df, min_data_size)
        if add_more_cols or _is_df_complete(df_append, df):
            df_result = df_append
    return df_result



def _setup_init_dfs(df: pd.DataFrame) -> tuple[pd.DataFrame, pd.DataFrame]:
    return pd.DataFrame(index=df.index), pd.DataFrame()


def _get_no_nan_rows(
    df: pd.DataFrame, df_no_nan: pd.DataFrame, remaining_cols: list[Any]
) -> list[int]:
    # creating this dataframe (with index only) is not necessary
    # but using it for concat instead of the whole df_no_nan
    # increases performance.
    df_dummy_no_nan = pd.DataFrame(index=df_no_nan.index)
    return [
        pd.concat(
            [df_dummy_no_nan, df[col].dropna()], axis=1, join="inner"
        ).shape[0]
        for col in remaining_cols
    ]


def _append_2_df_append(
    df_append: pd.DataFrame, df: pd.DataFrame, col_2_append: Any
) -> pd.DataFrame:
    return pd.concat(
        [df_append, df[col_2_append].dropna()], axis=1, join="inner"
    )


def _should_add_more_cols(
    df_no_nan: pd.DataFrame, df: pd.DataFrame, min_data_size: int | float
) -> bool:
    if df_no_nan.shape[1] == df.shape[1]:
        return False
    elif min_data_size < 1:
        return (df_no_nan.shape[0] / df.shape[0]) >= min_data_size
    else:
        return df_no_nan.shape[0] >= min_data_size


def _is_df_complete(df_append: pd.DataFrame, df: pd.DataFrame) -> bool:
    return (df_append.shape[1] == df.shape[1]) and (
        df_append.dropna().shape[0] > 0
    )
```

```python
from typing import Optional, Any, NoReturn

import pandas as pd

class DataCreator:
    def __init__(
        self,
        df_emacros: pd.DataFrame,
        nutrient_search: NutrientSearch,
        food_name_search: FoodNameSearch,
    ) -> None:
        self.df_emacros = df_emacros
        self.nutrient_search = nutrient_search
        self.food_name_search = food_name_search

        self.df_nutrients = nutrient_search.df_nutrients
        self.df_nutrients_details = nutrient_search.df_nutrients_details

    def create_data(
        self,
        nutrient_ids: list[int] | tuple[int, ...] | pd.Index,
        no_nans: bool = True,
        with_nutrient_names: bool = True,
        include_food_name: bool = True,
        shallow_valid_mass: bool = True,
        min_data_size: int | float = 0.1,
    ) -> pd.DataFrame:
        """
        - nutrient_ids: nutrients to include besides macros ans energy.
        - no_nans: whether to filter nans (maximan no nan df)
        - with_nutrient_names: whether to use nutrient names for columns
        -   instead of their ids.
        - include_food_name: wheter to include a column with the food name,
        -    thre might be duplicated names.
        - shallow_valid_mass: whether to drop foods with mass measurements
            (other than macros) that exceed 100g.
        - min_data_size: if int >= 1 then it is the minimum number of rows
        -   to keep while looking for no NaN rows
        -   if float < 1 then the minimum number or rows to keep is
        -   min_data_size * len(df)
        All data is per 100g of food.
        """
        _nutrient_ids = pd.Index(nutrient_ids)
        self._validate_nutrient_ids(_nutrient_ids)
        _df_nutrients = self._select_nutrients(
            _nutrient_ids, with_nutrient_names, shallow_valid_mass
        )
        df_result = self._concat_with_emacros(_df_nutrients)
        if no_nans:
            df_result = get_maximal_no_nan_df(df_result, min_data_size)
        if include_food_name:
            df_result = self._set_food_names(df_result)
        return df_result

    # ----------------------------------------------------------------

    def _validate_nutrient_ids(
        self, nutrient_ids: pd.Index
    ) -> Optional[NoReturn]:
        """
        raise ValueError if at least one of the nutrient_ids
        is not in df_nutrients.
        """
        if not nutrient_ids.isin(self.df_nutrients.columns).all():
            raise ValueError(
                "Incorrect nutrient_ids, check the ids and try again."
            )
        return None

    def _select_nutrients(
        self,
        nutrient_ids: pd.Index,
        with_nutrient_names: bool,
        shallow_valid_mass: bool,
    ) -> pd.DataFrame:
        _df_nutrients = self.df_nutrients[nutrient_ids]
        if shallow_valid_mass:
            _df_nutrients = self._drop_invalid_mass_foods(_df_nutrients)
        if with_nutrient_names:
            _df_nutrients = self._set_nutrient_names(_df_nutrients)
        return _df_nutrients

    def _drop_invalid_mass_foods(
        self, df_nutrients: pd.DataFrame
    ) -> pd.DataFrame:
        """df_nutrients columns are nutrient_ids."""
        mass_nutrient_ids = self.df_nutrients_details.loc[
            self.df_nutrients_details.unit_name == "G"
        ].unit_name.index
        mass_cols = df_nutrients.columns.intersection(mass_nutrient_ids)
        bool_invalid_s = (df_nutrients[mass_cols] > 100).any(axis=1)
        invalid_mass_fdc_ids = bool_invalid_s[bool_invalid_s].index
        return df_nutrients.drop(labels=invalid_mass_fdc_ids)

    def _set_nutrient_names(
        self, df_nutrient_cols: pd.DataFrame
    ) -> pd.DataFrame:
        """df_nutrient_cols columns are all nutrient_ids."""
        nutrient_ids = df_nutrient_cols.columns.intersection(
            self.df_nutrients_details.index
        )
        nutrient_names = self.df_nutrients_details.loc[nutrient_ids].name
        df_result = df_nutrient_cols.copy()[nutrient_ids]
        df_result.columns = nutrient_names
        return df_result

    def _concat_with_emacros(self, df_fdc_id) -> pd.DataFrame:
        """df_fdc_id has fdc_id index"""
        return pd.concat([self.df_emacros, df_fdc_id], axis=1, join="inner")

    def _set_food_names(self, df_fdc_idx: pd.DataFrame) -> pd.DataFrame:
        """
        df_fdc_idx index is fdc_id.
        """
        names = self.food_name_search.search_fdc_ids(df_fdc_idx.index)
        df_result = df_fdc_idx.copy().loc[names.index]
        df_result.insert(loc=0, column="food_name", value=names)
        return df_result
```

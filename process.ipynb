```python
PATH_TO_FOODDATA_CENTRAL_DIR = "./FoodData_Central_csv_2024-04-18"
PATH_TO_PROCESSED_DATA_DIR = "./processed_data"
```

```python
import pandas as pd
```

```python
raw_nutrient_file_name = "food_nutrient.csv"
raw_nutrient_path = f"{PATH_TO_FOODDATA_CENTRAL_DIR}/{raw_nutrient_file_name}"

df_nutrients_raw = pd.read_csv(raw_nutrient_path)
df_nutrients_raw
```

```python
cols = ["fdc_id", "nutrient_id", "amount"]

mi_nutrients = df_nutrients_raw[cols].groupby(cols[:-1]).agg(lambda x: x)
mi_nutrients
```

```python
df_nutrients = mi_nutrients.unstack(level=1).droplevel(0, axis=1)
df_nutrients
```
